```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
plt.style.use('seaborn-white')
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_csv("yellow_tripdata_2018-05.csv")
```

```python
df
```

```python
df = df.drop(['mta_tax', 'tolls_amount', 'improvement_surcharge', 'store_and_fwd_flag'], axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns={"tpep_pickup_datetime": "pickup_datetime", "tpep_dropoff_datetime": "dropoff_datetime"})
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
# Plot
plt.scatter(df['trip_distance'], df['total_amount'], alpha=0.5)
plt.title('Scatter plot pythonspot.com')
plt.xlabel('trip_distance')
plt.ylabel('total_amount')
plt.show()
```

```python
cleasing = df[(df.tip_amount >= 0) & (df.tip_amount <= 30) & 
              (df.fare_amount >= 2.5) & (df.fare_amount <= 100) & 
              (df.total_amount >= 2.5) & (df.total_amount <= 100) & 
              (df.trip_distance > 0) & (df.trip_distance <= 100) & # per mile
              (df.passenger_count > 0) &
              (df.extra >= 0) & (df.extra <= 1) &
              (df.RatecodeID >= 1) & (df.RatecodeID <= 6)]
```

```python
cleasing.describe()
```

```python
# Feature engineering
cleasing['pickup_datetime'] = pd.to_datetime(cleasing['pickup_datetime'], errors='coerce')
cleasing['year'] = cleasing.pickup_datetime.dt.year
cleasing['month'] = cleasing.pickup_datetime.dt.month
cleasing['day'] = cleasing.pickup_datetime.dt.day
cleasing['weekday'] = cleasing.pickup_datetime.dt.weekday
cleasing['hour'] = cleasing.pickup_datetime.dt.hour
```

```python
cleasing.describe()
```

```python
cleasing = cleasing.loc[cleasing['year'] == 2018]
cleasing = cleasing.loc[cleasing['month'] == 5]
```

```python
cleasing.describe()
```

```python
cleasing
```

```python
cleasing['pickup_datetime']= pd.to_datetime(cleasing['pickup_datetime'])
cleasing['dropoff_datetime'] = pd.to_datetime(cleasing['dropoff_datetime'])
cleasing['duration'] = cleasing['dropoff_datetime'] - cleasing['pickup_datetime']
cleasing['duration'] = cleasing['duration'].astype('timedelta64[m]')
```

```python
# Add external
cleasing['weather'] = 'sunny'
cleasing.weather[(cleasing['day'] == 5) | (cleasing['day'] == 4) | (cleasing['day'] == 16) | 
                (cleasing['day'] == 13)] = 'rainy' # cloud, light rain

cleasing.weather[(cleasing['day']== 30) | (cleasing['day'] == 29) | (cleasing['day'] == 28) | 
                (cleasing['day'] ==  9) | (cleasing['day'] ==  8)] = 'overcast' # scattered cloud

cleasing.weather[(cleasing['day'] == 21) | (cleasing['day'] == 22) | (cleasing['day'] == 14) | 
                (cleasing['day'] == 12) | (cleasing['day'] == 13) | (cleasing['day'] == 7) | (cleasing['day'] == 2)] = 'snow'

#cleasing.weather[cleasing['day'] in [7, 10, 12, 17]] = 'overcast'
```

```python
cleasing.groupby(['weather']).groups.keys()
```

```python
cleasing['weather_code'] = 1
cleasing.weather_code[cleasing['weather'] == 'rainy'] = 2
cleasing.weather_code[cleasing['weather'] == 'overcast'] = 3
cleasing.weather_code[cleasing['weather'] == 'snow'] = 4
```

```python
cleasing.groupby(['weather_code']).groups.keys()
```

```python
cleasing['weather_code'].value_counts()
```

```python
cleasing['hour_interval'] = ''
cleasing.hour_interval[(cleasing['hour'] == 0) | (cleasing['hour'] == 1) | (cleasing['hour'] == 2)] = '24-2'
cleasing.hour_interval[(cleasing['hour'] == 3) | (cleasing['hour'] == 4) | (cleasing['hour'] == 5)] = '3-5'
cleasing.hour_interval[(cleasing['hour'] == 6) | (cleasing['hour'] == 7) | (cleasing['hour'] == 8)] = '6-8'
cleasing.hour_interval[(cleasing['hour'] == 9) | (cleasing['hour'] == 10) | (cleasing['hour'] == 11)] = '9-11'
cleasing.hour_interval[(cleasing['hour'] == 12) | (cleasing['hour'] == 13) | (cleasing['hour'] == 14)] = '12-14'
cleasing.hour_interval[(cleasing['hour'] == 15) | (cleasing['hour'] == 16) | (cleasing['hour'] == 17)] = '15-18'
cleasing.hour_interval[(cleasing['hour'] == 18) | (cleasing['hour'] == 19) | (cleasing['hour'] == 20)] = '19-20'
cleasing.hour_interval[(cleasing['hour'] == 21) | (cleasing['hour'] == 22) | (cleasing['hour'] == 23)] = '21-23'
```
